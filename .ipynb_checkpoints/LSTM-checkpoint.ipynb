```python
import json
import joblib
import numpy as np
import pandas as pd
from numpy import array, hstack
from tensorflow import keras
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout, BatchNormalization, Lambda
from tensorflow.keras.regularizers import l1
```

## Data Preprocessing

```python
with open('balanced_dataset.json', 'r') as json_file:
    data = json.load(json_file)

def label_split_sequences(sequences, n_steps):
    y = list()
    for i in range(len(sequences)):
        end_ix = i + n_steps
        if end_ix >=len(sequences):
            break
        seq_y = sequences[end_ix]
        y.append(seq_y)
    return np.array(y)

def feature_split_sequences(sequences, n_steps):
    X = list()
    for i in range(len(sequences)):
        end_ix = i + n_steps
        if end_ix > len(sequences):
            break
        seq_x = sequences[i:end_ix]
        X.append(seq_x)
    return np.array(X)


label_encoder = LabelEncoder()
labels = [item['label'] for item in data]
norm = [np.mean(item['norm']) for item in data]
length = [len(item['norm']) for item in data]
range_list = [np.max(item['norm'],  axis=0) - np.min(item['norm'],  axis=0) for item in data]

# Fit the encoder on the labels and transform the labels to categorical values
categorical_labels = label_encoder.fit_transform(labels)


X1 = feature_split_sequences(categorical_labels, 3)
X2 = feature_split_sequences(np.array(length), 3)
X3 = feature_split_sequences(np.array(norm), 3)
X4 = feature_split_sequences(np.array(range_list), 3)
y = label_split_sequences(categorical_labels, 3)
X1 = np.array(X1[:-1])
X2 = np.array(X2[:-1])
X3 = np.array(X3[:-1])
X4 = np.array(X4[:-1])


all_labels = keras.utils.to_categorical(y, 8)
merged_features = np.stack((X1, X3, X2, X4), axis=2)

X_train, X_test, y_train, y_test = train_test_split(merged_features, all_labels, test_size=0.2, shuffle=True, random_state=42)
```

## Model Architecture

```python
model = Sequential()

model.add(LSTM(128, activation='relu', input_shape=(3, 4), return_sequences=True))
model.add(BatchNormalization())
model.add(Dropout(0.5))

model.add(LSTM(256))
model.add(BatchNormalization())
model.add(Dropout(0.5))

model.add(Dense(64, activation='relu', kernel_regularizer=l1(0.01)))
model.add(BatchNormalization())
model.add(Dropout(0.5))

model.add(Dense(8, activation='softmax', kernel_regularizer=l1(0.01)))

optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=2000, batch_size=64, validation_data=(X_test, y_test))
```
